range() 함수와 유사한 frange() 함수를 다음 조건을 참고하여 만드시오.  
1) frange()의 인자 구성은 다음과 같이 range와 동일하지만 각 인수들은 음수를 받지 않는다고 가정한다.
range(stop)  
range(start, stop[, step])    

2) frange() 함수의 인자에 대한 기본 시작(start) 값은 0.0이고, 기본 단계(step) 값은 0.1이다.  

3) frange 사용 예

    3-1) frange(0.5)  
        [0.0, 0.1, 0.2, 0.3, 0.4]  
    3-2) frange(1.0, 2.0)  
        [1.0, 1.1, 1.2, 1.3, 1.4, 1.5, 1.6, 1.7, 1.8, 1.9]  
    3-3) frange(2.2, 4.0, 0.5)  
        [2.2, 2.7, 3.2, 3.7]  

```python
def frange(stop,start=0,step=0.1):
    stop_real=int(stop*10)
    start_real=int(start*10)
    step_real=int(step*10)
    if start==0:
        return [round(i*0.1,1) for i in range(stop_real)]
    elif step==0.1:
        return [round(i*0.1,1) for i in range(stop_real,start_real)]
    else:
        return [round(i*0.1,1) for i in range(stop_real,start_real,step_real)]
print frange(0.6)
print frange(1.0,2.0)
print frange(2.2,4.0,0.5)
```

###해설)  
기본인수값을 사용하려하였지만 start의 위치로 인해서 조건문을 이용하여 인수가 들어오는 개수에 맞에 상황별로 인자 값을 삽입한다.
1. 인자가 하나인 경우는 자연스럽게 stop의 값을 입력하여 1/10의 값한 값을 반환
2. 인자가 2개인 것부터 start와 stop을 바꿔서 range값에 넣어준다.
3. 2와 마찬가지로 사용

    1) 다음에 제시되는 함수들을 모두 이용해야 한다.
        split
        map
        join
    2) myinitial() 함수 사용 예

```python
def myinitial(value):
    return "".join((map(lambda x:x[0].upper(),value.split())))
print myinitial("as soon as possible")
```

###[8번 문제] 프로젝트 오일러 문제 10  
10 이하의 소수를 모두 더하면 2 + 3 + 5 + 7 = 17 이 됩니다.

이백만(2,000,000) 이하 소수의 합은 얼마입니까?

```python
import math
def find_sosu(a):
    count=2
    while count<=math.sqrt(a):
        if a%count==0:
            return False
        count=count+1
    #print 'a : ',a    
    return True
print 2+sum(filter(lambda x:find_sosu(x),range(3,2000000,2)))
```

###해설)  
1. 소수를 판별하는 함수를 정의한다.
2. range 내장함수를 이요하여 1을 제외한 이백만까지의 홀수의 리스트를 생성한다.
3. filter()내장함수의 함수인자에 소수를 판별하는 객체와 홀수를 반환하는 리스트를 인자로 넣어 소수로 판별되면 합을 한 결과를 반환한다.

     1: 1  
     3: 1, 3

    10: 1, 2, 5, 10

```python
count=1# 맨처음 값은 1부터 시작
value=0# 판단할 변수
while 1:#정답을 찾을때까지 무한루프를 생성
    value+=count#순차적으로 값을 더한다.
    value_for_divide=2# 나누는 값은 2부터 시작
    dic={}#소인수분해를 저장할 변수객체 생성
    value_copy=value#오리지널값을 보존하기 위하여 값을 복사한다.
    while value_copy!=1:#소인수분해할 while루프문 생성
        if value_copy%value_for_divide==0:#만약 나누어지면 사전에 추가한다.
            value_copy/=value_for_divide
            if dic.get(value_for_divide):#기존에 존재하면 vlaue를 +1
                dic[value_for_divide]+=1
            else:#존재하지 않으면 value를 1로 설정하여 삽입
                dic[value_for_divide]=1
        else: #아니라면 나누는 값은 +1을하여 다음 값으로 나누어본다.
            value_for_divide+=1
    dic_values=dic.values()
    #reduce()내장함수를 사용하여 value리스트 내부의 값의 +1을 한 값들간의 곱을 출력한다.
    if (reduce(lambda x,y:x*(y+1),dic_values,1))>=500:
        print value,dic
        break
    count+=1            
        
```

###해설)

2. 무한루프를 생성하여 1,2,3,4,... 순으로 커지는 숫자의 값을 판단할 값에 더하는 설정을 한다.
3. 판단할 value의 값이 1이될 때까지 소인수분해를 한다.
4. 소인수분해한 값을 사전형 자료형 dic에 삽입한다.
5. dic에 vlaue를 추출하여 각 수의 1을 더한 값을 곱하여 500이 넘으면 value와 dic을 출력하며 while 무한루프를 멈춘다.

    4) 최소 5개 이상의 영어 웹 사이트에 대한 HTML 소스 파일과 단어 출현빈도 파일을 저장하시오.

        [주의] 영어 웹사이트 URL로만 5개 이상
    5) 위 문제에서 저장한 모든 words.frequency 파일들을 객체로 다시 로드하여 본인이 저장하여 분석한 사이트들에서 가장 많이 출현한 단어 3개를 뽑아 제시하시오.
        반드시 pickle 모듈로 저장한 5개 이상의 words.frequency를 다시 5개 이상의 사전 객체로 로드 하는 코드가 추가되어야 함


```python
import urllib2
import string
import pickle
#url의 소스를 읽어온다.
def read_source(address):
    return urllib2.urlopen(address).read()
#url의 소스를 이름에 맞추어 저장한다.
#이때, 앞의 http://를 제외한다.
def save_html(address):
    name=address
    if not name.find("http://"):
        name = name[name.find("http://")+7:]
    else:
        name = name
    name=name+".html.txt"
    f=open(name,"w")
    f.write(read_source(address))
    f.close()
#html,css,javascript등을 제거한다.    
def delete_tag(source):
    source1=source
    for i in range(0, source1.count('<script')):#javascript제거
        if source1.find("<script") != -1:
            source1 = source1.replace(source1[source1.find("<script"):source1.find("</script>")+9:],"")
    for i in range(0, source1.count('<style')):#css제거
        if source1.find("<style") != -1:
            source1 = source1.replace(source1[source1.find("<style"):source1.find("</style>")+8:],"")
    for i in range(0, source1.count('<!--')):#주석문제거
        if source1.find("<!--") != -1:
            source1 = source1.replace(source1[source1.find("<!--"):source1.find("-->")+3:],"")
    for i in range(0,source1.count("<div")):#<div>태그 제거
        if source1.find("<div") != -1:
            source1 = source1.replace(source1[source1.find("<div"):source1.find(">")+1:],"")
    for i in range(0,source1.count("<")):#남은 html제거
        if source1.find("<") != -1:
            source1 = source1.replace(source1[source1.find("<"):source1.find(">")+1:],"")
    return source1

#개행문자를 제거한다.
def delete_check1(source):
    source1=source
    check=list(string.punctuation)
    for i in check:
        source1=source1.replace(i,"")
    return source1

#영어 불용어를 제외하는데 tag를 제외할때와 다르게 추출된 사전형에서 pop()와 get()을 적절히 사용하여 제거한다.
#그 이유는 문자열에서 사용할 경우 apple은 a가 영어불용어에 해당되게 되어 pple와 같은 형태가 되기 때문이다.
def delete_check2(source_dic):#사전에서 pop을 사용하여 여기 있는 key값을 삭제할것이다.
    temp=source_dic
    check=[ 'a', 'and', 'are', 'as', 'at', 'be', 'but', 'by', 'for', \
        'if', 'in', 'into', 'is', 'it', 'no', 'not', 'of', 'on', 'or',\
        's', 'such', 't', 'that', 'the', 'their','then', 'there', 'these',\
        'they', 'this', 'to', 'was', 'will', 'with']
    for i in check:
        if temp.get(i):
            temp.pop(i)
    return temp

#문자열을 공백을 기준으로 분리한 뒤, 개수를 파악한다.
def count_word(source):
    source_list=source.split()
    source_dic={}
    for i in source_list:
        if not source_dic.get(i):
            source_dic[i]=source_list.count(i)     
    return source_dic
#사전형 정렬기준을 value로 설정한다.
def mycmp(a,b):
    return cmp(b[1],a[1])
#pickle모듈을 활용하여 정렬된 값을 저장한다.
def save_frequency(dic,address):
    name=address
    if not name.find("http://"):
        name = name[name.find("http://")+7:]
    else:
        name = name
    name=name+".word.frequency.txt"
    f=open(name,"w")
    pickle.dump(dic,f)
#pickle모듈을 활용하여 파일에 저장된 객체를 불러온다.
def load_frequency(address):
    name=address
    if not name.find("http://"):
        name = name[name.find("http://")+7:]
    else:
        name = name
    name=name+".word.frequency.txt"
    f=open(name)
    return pickle.load(f)
#위의 과정들을 종합적으로 수행하는 함수이다.
#즉, 함수를 부르는 함수이다.
def process(address):
    address_temp=address#변수를 보존하기 위해서 임의의 변수에 저장
    save_html(address_temp)#url 값의 소스를 받아온 뒤 .html로 저장한다.
    
    #1. 문자열을 부른다.
    #2. 주어진 영어 불용어는 다 소문자이기 때문에 값을 정렬할 때는 소문자로 계상하여야 편하기 때문에 소무자로 바꾼다.
    #3. javascript,css,html을 제거한다.
    #4. 각 단어의 개수를 파악한다.
    #5. 영어 불용어를 제거한다.
    #6. 단어의 빈도를 pickle을 이용하여 파일에 저장한다.
    save_frequency(\
    delete_check2(\
    count_word(\
    delete_check1(\
    delete_tag(\
    (read_source(address_temp)).lower() )))),address_temp)   
    #7. 파일에 있는 객체를 불러와 정렬을 한 뒤, 반환한다.
    loaded_list=(load_frequency(address_temp)).items()
    loaded_list.sort(mycmp)#사전형을 리스트형으로 반환하여 저장
    return loaded_list#사전형 반환
#--------------------------------------sart---------------------------------------------------------
#url을 리스트형식으로 가지고 있는 변수를 선언한 뒤 값을 append를 이용하여 저장한다.
address_group=[]
address_group.append("http://edition.cnn.com")
address_group.append("http://mashable.com")
address_group.append("http://www.amazon.com")
address_group.append("http://www.nytimes.com")
address_group.append("http://thenextweb.com")
#각 url별로 3개씩 최대 빈출 단어를 저장할 dic_ranking
dic_ranking={}
#address_group에 있는 값을 순서대로 process()모듈에 넣어 최대빈출이 정렬된 객체를 반환하게 한다.
for address in address_group:
    dic=process(address)
    count=0
    #각 단어의 최대빈출을 address_group에 넣고 기존에 존재한다면 크기비교를 하여 넣는다.
    for key,value in dic:
        if not dic_ranking.get(key):#없다면 삽입
              dic_ranking[key]=value
        else:#있다면 크기 비교
            if value>dic_ranking[key]:#기존값보다 크다면 변경
                dic_ranking[key]=value
        count+=1
        if count>=3:
            break;
#각각 3개씩 단어를 넣은 뒤 정렬을 하여 최대 rank 1, 2, 3을 출력하며 프로그램을 마친다.            
result=sorted(dic_ranking.items(),mycmp)[0:3]
for i,j in result:
    print unicode(i,'utf-8'),':',j
print
#print '-----------------------------------------------------------------------------------------------------------------'
```
