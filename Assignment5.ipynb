다음 각 요구사항 모두를 만족시키는 Counter 클래스를 코딩하시오 (정답을 각 요구사항별로 입력할 필요 없이 3번 문제에 대해 1개의 클래스 정의 코드를 제시하면 된다.)

요구사항 1. 생성자에 count 값과 step 값을 인자로 받을 수 있다.
        count: Counter 인스턴스가 지니는 초기 정수 값
        step: Counter 인스턴스의 count가 증가되는 증분 (defalt 값: 1)

요구사항 2. 다음과 같이 Counter의 인스턴스를 출력을 해주는 __str__() 메소드를 Counter 클래스 내에 구현하시오.

        >>> print c
        [Count (step: 1)] 10

        [Count (step: 2)] 10

요구사항 3. 다음과 같이 step에 주어진 증분만큼 count를 증가시키는 incr() 메소드를 Counter 클래스 내에 구현하시오.

        >>> c.incr()
        >>> d.incr()
        >>> print c
        [Count (step: 1)] 11
        >>> print d
        [Count (step: 2)] 12

요구사항 4. Counter 클래스 내에 관련 메소드를 추가하여 인스턴스 객체를 직접 호출(call)할 수 있도록 하시오. 인스턴스 객체를 직접 호출했을 때에 내부적으로 incr() 메소드를 호출하는 방법으로 구현하시오.

        >>> c()
        >>> d()
        >>> print c
        [Count (step: 1)] 12
        >>> print d
        [Count (step: 2)] 14


요구사항 5. 다음과 같은 두 개의 산술 연산 (+, -)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.

        >>> c = c + 5
        >>> d = d - 5
        >>> print c
        [Count (step: 1)] 17
        >>> print d
        [Count (step: 2)] 9

요구사항 6. 다음과 같은 관계연산 (>, <, ==)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.

        >>> print c > 10
        True
        >>> print d > 10
        False
        >>> print c < 10
        False
        >>> print d < 10
        True
        >>> print c == 17
        True
        >>> print d != 9
        False 

```python
class Counter:
    def __init__(self,count,step=1):#요구사항 1
        self.count=count
        self.step=step
    def __str__(self):#요구사항 2
        return "[Count (step: %d)] %d"%(self.step,self.count)
    def incr(self):#요구사항 3
        self.count=self.count+self.step
    def __call__(self):#요구사항 4
        self.incr()
    def __add__(self,other):#요구사항 5
        self.count = self.count + other
        return self
    def __sub__(self,other):#요구사항 5
        self.count = self.count - other
        return self
    def __gt__(self,other):#요구사항 6
        return self.count > other
    def __lt__(self,other):#요구사항 6
        return self.count < other
    def __eq__(self,other):#요구사항 6
        return self.count==other
    def __ne__(self,other):#요구사항 6
        return self.count!=other

#요구사항 1 ,2수행
c = Counter(10)
d = Counter(10, 2)
print "-------------print c, print d-------------"
print c
print d
print 
#요구사항 3 수행
c.incr()
d.incr()
print "-------------incr() 후 print c, print d----"
print c
print d
print 
#요구사항 4 수행
c()
d()
print "-------------호출(call) 후 print c, print d-------------"
print c
print d
print 
#요구사항 5 수행
c = c + 5
d = d - 5
print "-------------산술연산 (+, -)이 수행 후 print c, print d-------------"
print c
print d
print 
#요구사항 6 수행
print "-------------관계연산 (>, <, ==)이 수행 후 print c, print d-------------"
print "c > 10 : ",c > 10 
print "d > 10 : ",d > 10
print "c < 10 : ",c < 10
print "d < 10 : ",d < 10
print "c == 17 : ",c == 17
print "d != 9 : ",d != 9
```

4번 문제에 정의된 MySet 클래스에 메소드를 추가하여 다음 각 요구사항 모두를 만족시키는 코딩을 제시하시오

    정답을 각 요구사항별로 입력할 필요 없이 요구사항 3개 전체에 대해 1개의 MySet 클래스 정의 코드를 제시하고 주석등으로 해당 코드를 설명한다.

        >>> u = s | t
        >>> print u
        MySet: {1, 2, 3, 4, 5, 6, 7, 8, 9}

        >>> u = s & t

        >>> s = MySet([1, 2, 3])

        >>> u = s - t
        >>> print u

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result
    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)
            
    #요구사항 1        
    def __or__(self,other):# | 연산자를 오버로딩한다.
        r=[]#공백 리스트를 생성한다.
        for i in self:#self의 원소를 r에 삽입한다.
            r.append(i)    
        for i in other:#other의 원소가 r에 속해있지 않으면 r에 삽입한다.
            if i not in r:
                r.append(i)
        return MySet(r)#r을 반환한다.
    #요구사항 1
    def __and__(self,other):
        r=[]#공통원소를 넣을 리스트 생성
        for i in other:# other의 원소가 self에 있으면 r에 원소를 추가한다.
            if i in self:
                r.append(i)
        return MySet(r)#r을 반환한다.
    #요구사항 1
    def __sub__(self, other):
        for i in other:
            if i in self:
                self.pop(self.index(i))
        return self
s = MySet([1, 2, 2, 3])    
t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])    
u=s|t
print u# s|t의 결과를 통해 __or__이 올바르게 작동하는지 확인할 수 있다.
u=s&t
print u# s&t의 결과를 통해 __or__이 올바르게 작동하는지 확인할 수 있다.
u=s-t
print u# s-t의 결과를 통해 __or__이 올바르게 작동하는지 확인할 수 있다.
```

```python
import urllib2
import string
import StringIO
import sys
is_real=False#__main__():같은 역활을 하는 것으로 테스트할 때 로그를 찍기위해서 사용함
#사전형 정렬기준을 value로 설정한다.
class WebWordsFrequency:
    def __init__(self,*arg):#들어오는 주소들은 address_list 리스트에 저장한다.
        self.max=0
        self.address_list=[]#생성자 인자로 들어오는 주소들을 저장하는 변수
        self.word_frequency=[]#(단어 : 횟수)로 저장하는 리스트
        for address in arg:
            self.address_list.append(address)
    def addUrl(self,add_address):#문자열이 들어올 경우만 list에 추가한다.
        if type(add_address)==type(""):
            self.address_list.append(add_address)
    def removeUrl(self,remove_address):#값이 없을 경우 에러가 발생하므로 에러구문을 잘 처리해준다.
        try:
            self.address_list.remove(remove_address)
        except ValueError:
            print "해당값이 존재하지 않습니다."
    def listUrls(self):#리스트 목록을 출력해준다.
        for address in self.address_list:
            print address
    def read_source(self):#여러개의 사이트에 대한 종합적인 단어횟수 출력이기 때문에 여러개의 사이트의 url로부터 오는 문자열을 모두 합한다.
        #이때 사용하는 방법은 StringIO모듈을 사용하여 약간이라도 효율적으로 문자열을 연결한다.
        if is_real:
            print  "read_source :"
        stdout=sys.stdout
        sys.stdout=f=StringIO.StringIO()
        for address in self.address_list:
            print urllib2.urlopen(address).read()
        sys.stdout=stdout    
        return f.getvalue()
        
    #html,css,javascript등을 제거한다.    
    def delete_tag(self,source):
        if is_real:
            print  "delete_tag :"
        source1=source
        for i in range(0, source1.count('<script')):#javascript제거
            if source1.find("<script") != -1:
                source1 = source1.replace(source1[source1.find("<script"):source1.find("</script>")+9:],"")
        for i in range(0, source1.count('<style')):#css제거
            if source1.find("<style") != -1:
                source1 = source1.replace(source1[source1.find("<style"):source1.find("</style>")+8:],"")
        for i in range(0, source1.count('<!--')):#주석문제거
            if source1.find("<!--") != -1:
                source1 = source1.replace(source1[source1.find("<!--"):source1.find("-->")+3:],"")
        for i in range(0,source1.count("<div")):#<div>태그 제거
            if source1.find("<div") != -1:
                source1 = source1.replace(source1[source1.find("<div"):source1.find(">")+1:],"")
        for i in range(0,source1.count("<")):#남은 html제거
            if source1.find("<") != -1:
                source1 = source1.replace(source1[source1.find("<"):source1.find(">")+1:],"")
        return source1

    #개행문자를 제거한다.
    def delete_check1(self,source):
        if is_real:
            print  "delete_check1 :"
        source1=source
        check=list(string.punctuation)
        for i in check:
            source1=source1.replace(i,"")
        return source1

    #영어 불용어를 제외하는데 tag를 제외할때와 다르게 추출된 사전형에서 pop()와 get()을 적절히 사용하여 제거한다.
    #그 이유는 문자열에서 사용할 경우 apple은 a가 영어불용어에 해당되게 되어 pple와 같은 형태가 되기 때문이다.
    def delete_check2(self,source_dic):#사전에서 pop을 사용하여 여기 있는 key값을 삭제할것이다.
        if is_real:
            print  "delete_check2 :"
        temp=source_dic
        check=[ 'a', 'and', 'are', 'as', 'at', 'be', 'but', 'by', 'for', \
            'if', 'in', 'into', 'is', 'it', 'no', 'not', 'of', 'on', 'or',\
            's', 'such', 't', 'that', 'the', 'their','then', 'there', 'these',\
            'they', 'this', 'to', 'was', 'will', 'with']
        for i in check:
            if temp.get(i):
                temp.pop(i)
        return temp

    #문자열을 공백을 기준으로 분리한 뒤, 개수를 파악한다.
    def count_word(self,source):
        if is_real:
            print  "count_word :"
        source_list=source.split()
        source_dic={}
        for i in source_list:
            if not source_dic.get(i):
                source_dic[i]=source_list.count(i)     
        return source_dic

    #위의 과정들을 종합적으로 수행하는 함수이다.
    #즉, 함수를 부르는 함수이다.
    def process(self):
        if is_real:
            print  "process :"
        #1. 문자열을 부른다.
        #2. 주어진 영어 불용어는 다 소문자이기 때문에 값을 정렬할 때는 소문자로 계상하여야 편하기 때문에 소무자로 바꾼다.
        #3. javascript,css,html을 제거한다.
        #4. 각 단어의 개수를 파악한다.
        #5. 영어 불용어를 제거한다.
        #6. 빈출횟수와 단어가 한쌍인 사전형 반환을 한다.
        dic_temp=self.delete_check2(self.count_word(self.delete_check1(self.delete_tag((self.read_source()).lower() ))))   
        #7. 사전형을 튜플(단어 : 개수)형태가 하나의 인자인 list로 저장한다.
        list_temp=dic_temp.items()
        self.word_frequency=list_temp
        return self.word_frequency
    def getWordsFrequency(self):
        #주소가 없는 경우는 None을 반환
        if len(self.address_list)==0:
            return {}
        #주소가 있는 경우 process()모듈을 실행
        else:
            return self.process()
    #최대 횟수 단어 출력함수
    def getMaxFreqencyWords(self):
        if len(self.word_frequency)==0:#기존에 분석한 자료가 없다면 새롭게 분석 시작
            self.getWordsFrequency()
        max=0#최대횟수 저장 변수
        max_list=[]#최대 횐수 단어저장 리스트 - 복수개일 수 있기때문에 리스트형태로 만든다.
        #한번의 탐색으로 최대횟수 리스트를 출력하기 위해 최대값이 갱신될 때마다 거기에 맞는 단어를 리스트에 저장한다.
        for i in self.word_frequency:
            if i[1]>max:
                max=i[1]
                max_list=[i[0]]
            elif i[1]==max:
                max_list.append(i[0])
        #최대 횟수의 단어를 출력한다.        
        for word in max_list:
            print word
#생성자로 주소 삽입 구문            
a=WebWordsFrequency('http://www.times.com','https://www.amazon.com')
#addUrl()사용 주소 리스트 추가
a.addUrl('https://github.com')
a.addUrl("http://www.google.com")
#removeUrl()이용 주소 삭제 
a.removeUrl("http://www.google.com")
print "*"*80
print "listUrls() :"
#저장된 주소 리스트 출력
a.listUrls()
print "*"*80
print "getWordsFrequency() : "
#단어 : 횟수 가 저장된 리스트 출력
print a.getWordsFrequency()
print
print "*"*80
print "getMaxFreqencyWords() : ",
#최대 횟수 단어 출력
a.getMaxFreqencyWords()
```
